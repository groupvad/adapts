```python
import torch

def get_model_size_mb(model):
    # Calcola il numero totale di byte
    param_size = 0
    for param in model.parameters():
        param_size += param.nelement() * param.element_size()
    
    # Calcola la dimensione dei buffer (es. running mean/var di BatchNorm)
    buffer_size = 0
    for buffer in model.buffers():
        buffer_size += buffer.nelement() * buffer.element_size()

    size_all_mb = (param_size + buffer_size) / 1024**2
    return size_all_mb

# Esempio d'uso:
# my_model = MyModule()
# print(f"Dimensione modello: {get_model_size_mb(my_model):.2f} MB")
```
